```python
import os
save = []
for root, dirs, files in os.walk("models", topdown=False):
    root_in_save = False
    for s in save:
        if root in s:
            root_in_save = True
    if root_in_save:
        continue
    if root.endswith('generator') or root.endswith('discriminator'):
        bins_files = []
        for file in files:
            if file.endswith('.bin') or file.endswith('.pt'):
                to_remove = os.path.join(root, file)
                bins_files.append(to_remove)
                #os.remove(to_remove)
        bins_files.sort()
        for file in bins_files[:-2]:
            to_remove = file
            print(to_remove)
            os.remove(to_remove)
```
