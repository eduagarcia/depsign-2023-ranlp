```python
import os
import pandas as pd
```

```python
output_path = 'kfold'
os.makedirs(output_path, exist_ok=True)
```

```python
train_file = '../data/train_initial.csv'
dev_file = '../data/dev_data.csv'

train_df = pd.read_csv(train_file)
dev_df = pd.read_csv(dev_file)
```

```python
train_df
```

```python
dev_df
```

```python
all_df = pd.concat((train_df, dev_df), ignore_index=True)
```

```python
all_df
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=4, shuffle=True, random_state=42)
for i, (train_index, test_index) in enumerate(kf.split(all_df)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}, len={len(train_index)}")
    print(f"  Test:  index={test_index}, len={len(test_index)}")
    k_train_df = all_df.loc[train_index]
    k_test_df = all_df.loc[test_index]
    k_output_path = os.path.join(output_path, str(i))
    os.makedirs(k_output_path, exist_ok=True)
    #k_train_df.to_csv(os.path.join(k_output_path, 'train_initial.csv'), index=False)
    #k_test_df.to_csv(os.path.join(k_output_path, 'dev_data.csv'), index=False)
```
